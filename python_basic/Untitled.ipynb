```python
import requests
from bs4 import BeautifulSoup
url = 'https://finance.naver.com/sise/sise_quant.naver'
```

```python
res = requests.get(url).text
```

```python
soup = BeautifulSoup(res,'lxml')
result = soup.select('.tltle')
```

```python
name=[]
for item in result:
    name.append(item.string.strip())
```

```python
name
```

```python
result = soup.select('td.number')
```

```python
result
```

```python
now_price = []
count=1
for item in result:
    if count%10==1:
        now_price.append(float(item.string.replace(',','')))
    count=count+1
```

```python
now_price
```

```python
list(zip(name,now_price))
```
